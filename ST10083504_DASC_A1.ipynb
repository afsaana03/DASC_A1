The pairplot visually represents these relationships by plotting each of the selected musical attributes against the others, allowing for the observation of patterns specific to each genre (Geeksforgeeks, 2024). The hue parameter (playlist_genre) highlights how different genres cluster around certain attribute values, making it easier to see trends like those described above (Geeksforgeeks, 2024).

This kind of analysis is valuable for understanding the distinct characteristics of various music genres and how they relate to listener experiences, such as energy levels and danceability, which can influence mood and activity (Geeksforgeeks, 2024). It also provides insights into how specific attributes are distributed within and across genres, which can be useful for tasks like music recommendation or playlist curation (Geeksforgeeks, 2024).

```python
selected_columns = ['energy', 'danceability', 'valence', 'playlist_genre']
```

```python
sns.pairplot(cleaned_data[selected_columns], hue='playlist_genre')
plt.show()
```

Observations from the Pairplot:
EDM Songs:

Energy and Valence: Songs in the EDM (Electronic Dance Music) genre tend to be more energetic and possess a higher valence. Valence is a measure of the musical positivity conveyed by a track, with higher values indicating more positive or upbeat songs. The combination of high energy and high valence suggests that EDM tracks are typically lively, intense, and uplifting, which aligns with the genre's characteristic high-energy beats and vibrant atmosphere.
R&B Genre:

Energy Levels: On the other hand, R&B (Rhythm and Blues) tracks generally exhibit lower energy levels compared to EDM. R&B music often emphasizes smooth, soulful melodies and emotional depth, which may explain the lower energy observed in these tracks. This genre is more likely to feature slower tempos and softer sounds, contributing to a more relaxed and mellow listening experience.
Latin Genre:

Danceability: The Latin genre stands out as being particularly suitable for dancing, as indicated by higher danceability scores. Danceability describes how suitable a track is for dancing, and the Latin genre, with its rhythmic beats and lively tempos, naturally encourages movement. This aligns with the cultural significance of dance in Latin music, where rhythms like salsa, reggaeton, and samba are central to the genre's identity.
Rap Songs:

Diverse Energy and Danceability: Rap songs in the dataset show a wide range of energy levels, with some tracks being very low in energy while others are quite high. This diversity reflects the broad spectrum of styles within the rap genre, from slow, lyrical tracks to fast-paced, high-energy anthems. Additionally, some rap songs exhibit high danceability and valence, suggesting that while they might maintain the rhythmic and lyrical complexity typical of rap, they can also be upbeat and dance-friendly.

Pie chart: different genres (Geeksforgeeks, 2024).

```python

custom_colors = ['purple', 'lightblue', 'green', 'pink', 'orange','lightgreen',]
paper = plt.figure(figsize=[10,6])
cleaned_data['playlist_genre'].value_counts().plot.pie(autopct='%3.1f%%',colors=custom_colors)
plt.title('Genre')
plt.show()
```

This is using Matplotlib to create a bar chart that visualizes the top 10 most common playlist_subgenre values from the cleaned_data dataset (Geeksforgeeks, 2024).

```python
cleaned_data['playlist_subgenre'].value_counts()
paper = plt.figure(figsize=(15,7))

cleaned_data['playlist_subgenre'].value_counts()[:10].plot.bar(color='lightblue')
plt.title('playlist_subgenre',size=14)
plt.xlabel('palylist_subgenre',size=12)
plt.ylabel('Count',size=12)
plt.xticks(rotation=45)
plt.show()
```

```python
print(cleaned_data['album_release_year'].max())
print(cleaned_data['album_release_year'].min())

# year with releases
cleaned_data['album_release_year'].value_counts()
```

This generates a bar chart that visualizes the top 10 years with the highest number of album releases from the cleaned_data dataset (Geeksforgeeks, 2024).

```python
paper = plt.figure(figsize=(15,7))
cleaned_data['album_release_year'].value_counts()[:10].sort_values(ascending=False).plot.bar(edgecolor='red', color='yellow')
plt.xlabel('Year',size=15)
plt.ylabel('Count',size=15)
plt.title('No. of Album Release Year wise',size=17)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
